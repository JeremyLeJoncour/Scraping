```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:root@localhost:3307/scraping")
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data = pd.read_sql_query("SELECT time, value, percent FROM enjin", engine)
data = data[337:]

fig = make_subplots(rows=1, cols=2)

fig.add_trace(go.Scatter(x=data['time'], y=data['value']), row=1, col=1)
fig.update_yaxes(title_text="Valeur ($)", row=1, col=1)
fig.update_xaxes(title_text='Temps')

fig.add_trace(go.Scatter(x=data['time'], y=data['percent']), row=1, col=2)
fig.update_yaxes(title_text="Pourcentage (%)", row=1, col=2)
fig.update_xaxes(title_text='Temps')

fig.update_layout(title='Evolution du Enjin Coin')


fig.show()
```
