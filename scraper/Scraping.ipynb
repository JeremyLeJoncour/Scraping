```python
from selenium import webdriver
from datetime import datetime
import time
import pandas as pd
```

```python
import mysql.connector

config = {
    'user': 'root',
    'password': 'root',
    'host': 'localhost',
    'port': '3307',
    'database': 'scraping',
    'raise_on_warnings': True,
}

#CrÃ©ation du Cursor
link = mysql.connector.connect(**config)
cursor = link.cursor()
```

```python
driver = webdriver.Chrome('C:/Users/utilisateur/chromedriver.exe')
```

```python
driver.get('https://coincap.io/assets/enjin-coin')
```

```python
list_value = []
list_percent = []
list_time = []
```

```python
timed_start = datetime.now().minute + datetime.now().hour*60
print(timed_start)
```

```python
pull = 60
timed_end = datetime.now().minute + datetime.now().hour*60 + pull
print(timed_end)
```

```python
while timed_start < timed_end :
    
    value = driver.find_element_by_xpath('/html/body/div[2]/main/div[2]/div/div[1]/div/div[2]/h2/span').text
    value = float(value.replace('$', ''))
    
    percentage = driver.find_element_by_xpath('/html/body/div[2]/main/div[2]/div/div[1]/div/div[2]/h3/span').text
    percentage = float(percentage.replace('%', ''))
    
    now = datetime.now()
    timed_start = now.minute + now.hour*60
    
    list_value.append(value)
    list_percent.append(percentage)
    list_time.append(f'{now.hour:02d}:{now.minute:02d}')
    cursor.execute(f"INSERT INTO enjin (`id`, `time`, `value`, `percent`) VALUES (NULL, '{now}', {value}, {percentage})")
    link.commit()

    print(f'{now.day:02d} {now.strftime("%B")} {now.hour:02d}:{now.minute:02d} \t==>\t {value} \t==>\t {percentage}')
    
    data = {'Time': list_time, 'Value ($)': list_value, 'Progress (%)': list_percent}
    data = pd.DataFrame(data, columns = ['Time', 'Value ($)', 'Progress (%)'])
    data.to_csv('data.csv', index=False)
    
    time.sleep(15)
```
